```python
from itertools import combinations
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

data = df[df['Order ID'].duplicated(keep=False)]
data['Combine'] = df.groupby('Order ID',as_index=False)['Product'].transform(lambda x: ','.join(x))

pd.set_option('mode.chained_assignment', None)
data.head()
```

```python
data = data[['Order ID', 'Combine']].drop_duplicates()
count = Counter()
for row in data['Combine']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key, value in count.most_common(10):
    print(key, value)
```

```python
# x input
feature_cols=['Quantity Ordered','Price Each']
x=df[feature_cols]
x.head()
```

```python
# y output
y=df['Sales Revenue']
y.head()
```

```python
# Spliting x and y into training and testing set
from sklearn.linear_model import LinearRegression
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=1)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#x_test.reset_index(drop=True)
x_test=pd.DataFrame(x_test)
x_test.head()
```

```python
y_test.reset_index(drop=True)
```

```python
# model fitting 
linreg=LinearRegression()
linreg.fit(x_train, y_train)
```

```python
# interpreting a model coefficient 
print(linreg.intercept_)
print(linreg.coef_)
```

```python
list(zip(feature_cols, linreg.coef_))
```

```python
predicted=linreg.predict(x_test)
print(predicted)
```

```python
dfr=pd.DataFrame({'Actual':y_test, 'Predicted':predicted})
dfr.head(20)
```
